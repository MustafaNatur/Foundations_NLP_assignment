### 4.2 Loss curves

Create a **Python script** that:

• Parses the training logs for each run (nanoGPT prints step, train loss, val loss to stdout).

• Plots **training loss vs. step** for all runs in a single figure (one line per run, with a legend).

• Plots **validation loss vs. step** in a second figure.

Save these as .png or .pdf files. **Tip:** Redirect output to a log file during training:

```python
!python train.py config/train_shakespeare_char_baseline.py | tee logs/baseline.log
```

```python
# OUR PYTHON CODE
from pathlib import Path
from loss_curves import parse_log_file, plot_training_loss, plot_validation_loss

LOG_FILES = [
    "logs/baseline.log",
    "logs/low_lr.log",
    "logs/2_layers.log",
    "logs/4_layers.log",
    "logs/128_width.log",
    "logs/256_width.log",
    "logs/64_context.log",
    "logs/128_context.log",
    "logs/1000_iters.log",
    "logs/2500_iters.log",
]

all_runs = {}

for log_name in LOG_FILES:
    log_path = Path(log_name)

    if not log_path.exists():
        print(f"Skipping missing file: {log_path}")
        continue

    steps, train_losses, val_losses = parse_log_file(log_path)

    run_name = log_path.stem
    all_runs[run_name] = {
        "steps": steps,
        "train": train_losses,
        "val": val_losses,
    }
    
if not all_runs:
    raise FileNotFoundError("No valid log files were found.")

plot_training_loss(all_runs)
plot_validation_loss(all_runs)
```
