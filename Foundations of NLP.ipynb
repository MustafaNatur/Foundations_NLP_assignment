## 4. Evaluation and Reporting

4.1 Quantitative evaluation

For each run, record:

• Final training loss

• Final validation loss

• Training time (wall-clock)

```python
def readOutLogFile(folderPath, fileName):
    import os
    import numpy as np
    from string import printable

    logFileContent = ""

    filePath = os.path.join(folderPath, fileName)

    with open(filePath, 'r') as logFile:

        logFileContent = logFile.read()

    logData = {}
    logData["path"] = filePath
    logData["Experiment"] = fileName.replace("_", " ").replace(".log", "")
    logData["Iters"] = 5000
    
    itterationTimes = 0
    evaluationTimes = 0
    eval_interval = 250
    lastIteration = 0

    lines = logFileContent.split("\n")
    for line in lines:
        lineContext = ''.join(char for char in line if char in printable)
        lineContext = lineContext.replace(" ", "").replace("#CHANGEFROMBASELINE", "")
        if len(lineContext) > 1:
            if lineContext.startswith("learning_rate"):
                logData["LR"] = f"{float(lineContext.replace("learning_rate=", "")):.1e}"
            if lineContext.startswith("n_layer"):
                logData["Layers"] = int(lineContext.replace("n_layer=", ""))
            if lineContext.startswith("n_embd"):
                logData["Embd"] = int(lineContext.replace("n_embd=", ""))
            if lineContext.startswith("n_head"):
                logData["n_head"] = int(lineContext.replace("n_head=", ""))
            if lineContext.startswith("block_size"):
                logData["Block"] = int(lineContext.replace("block_size=", ""))
            if lineContext.startswith("dropout"):
                logData["Dropout"] = float(lineContext.replace("dropout=", ""))
            if lineContext.startswith("max_iters"):
                logData["Iters"] = int(lineContext.replace("max_iters=", ""))
            if lineContext.startswith("eval_interval"):
                eval_interval = int(lineContext.replace("eval_interval=", ""))
            if lineContext.startswith("step"):

                stepData = {}
                
                lineContextParts = lineContext.replace(":", ",").split(",")

                stepData["step"] = int(lineContextParts[0].replace("step", ""))
                stepData["trainloss"] = float(lineContextParts[1].replace("trainloss", ""))
                stepData["valloss"] = float(lineContextParts[2].replace("valloss", ""))

                if not ("steps" in logData):
                    logData["steps"] = []
                
                logData["steps"].append(stepData)
            if lineContext.startswith("iter"):

                iterData = {}
                
                lineContextParts = lineContext.replace(":", ",").split(",")

                iterData["iter"] = int(lineContextParts[0].replace("iter", ""))
                iterData["loss"] = float(lineContextParts[1].replace("loss", ""))
                iterData["time"] = float(lineContextParts[2].replace("time", "").replace("ms", ""))
                iterData["mfu"] = float(lineContextParts[3].replace("mfu", "").replace("%", ""))

                if not ("iter" in logData):
                    logData["iter"] = []

                if not (iterData["iter"] % eval_interval == 0):
                    itterationTimes += iterData["time"] * (iterData["iter"] - lastIteration)
                else:
                    evaluationTimes += iterData["time"]
                    if len(logData["iter"]) > 0 and iterData["iter"] - lastIteration - 1 >= 1:
                        itterationTimes += logData["iter"][-1]["time"] * (iterData["iter"] - lastIteration - 1)
                    
                logData["iter"].append(iterData)
                lastIteration = iterData["iter"]

    logData["Time (min)"] = round((itterationTimes + evaluationTimes) / 1000 / 60, 2)
    logData["Train Loss"] = logData["steps"][-1]["trainloss"]
    logData["Val Loss"] = logData["steps"][-1]["valloss"]

    return logData
```

```python
logsFolderPath = "logs/"
logFileNames = ["baseline.log", "low_lr.log", "2_layers.log", "4_layers.log", "64_context.log", "128_context.log", "128_width.log", "256_width.log", "2500_iters.log", "1000_iters.log"]
tableColumns = ["Experiment", "LR", "Layers", "Embd", "Dropout", "Iters", "Train Loss", "Val Loss", "Time (min)"]

headderStr = "|"
seperatorStr = "|"
for column in tableColumns:
    headderStr += " " + "{:<11}".format(column) + " |"
    seperatorStr += "-" * 13 + "|"

print(headderStr)
print(seperatorStr)

for logFileName in logFileNames:
    logFileData = readOutLogFile(logsFolderPath, logFileName)
    
    rowStr = "|"

    for column in tableColumns:
        if column in logFileData:
            rowStr += " " + "{:<11}".format(str(logFileData[column])) + " |"
        else:
            rowStr += " |"

    print(rowStr)
```
