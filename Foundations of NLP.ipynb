```python
# Parse final metrics from all Pride and Prejudice logs and compare them with the Shakespeare 256-width run
import sys
import importlib
from pathlib import Path

sys.path.append("..")
import loss_curves
importlib.reload(loss_curves)

from loss_curves import parse_log_file, parse_runtime_minutes

pride_prejudice_run_specs = {
    "Pride_and_Prejudice_Baseline": {
        "log": "../logs/pride_prejudice_baseline.log",
        "out_dir": "../Weights/out-pride-prejudice-baseline",
    },
    "Pride_and_Prejudice_Low_LR": {
        "log": "../logs/pride_prejudice_low_lr.log",
        "out_dir": "../Weights/out-pride-prejudice-low_lr",
    },
    "Pride_and_Prejudice_2_Layers": {
        "log": "../logs/pride_prejudice_2_layers.log",
        "out_dir": "../Weights/out-pride-prejudice-2_layers",
    },
    "Pride_and_Prejudice_4_Layers": {
        "log": "../logs/pride_prejudice_4_layers.log",
        "out_dir": "../Weights/out-pride-prejudice-4_layers",
    },
    "Pride_and_Prejudice_64_Context": {
        "log": "../logs/pride_prejudice_64_context.log",
        "out_dir": "../Weights/out-pride-prejudice-64_context",
    },
    "Pride_and_Prejudice_128_Context": {
        "log": "../logs/pride_prejudice_128_context.log",
        "out_dir": "../Weights/out-pride-prejudice-128_context",
    },
    "Pride_and_Prejudice_128_Width": {
        "log": "../logs/pride_prejudice_128_width.log",
        "out_dir": "../Weights/out-pride-prejudice-128_width",
    },
    "Pride_and_Prejudice_256_Width": {
        "log": "../logs/pride_prejudice.log",
        "out_dir": "../Weights/out-pride-prejudice",
    },
    "Pride_and_Prejudice_1000_Iters": {
        "log": "../logs/pride_prejudice_1000_iters.log",
        "out_dir": "../Weights/out-pride-prejudice-1000_iters",
    },
    "Pride_and_Prejudice_2500_Iters": {
        "log": "../logs/pride_prejudice_2500_iters.log",
        "out_dir": "../Weights/out-pride-prejudice-2500_iters",
    },
}


def final_metrics(log_name):
    log_path = Path(log_name)
    steps, train_losses, val_losses = parse_log_file(log_path)
    if not steps:
        raise ValueError(f"No step metrics found in {log_path}")
    return {
        "final_step": steps[-1],
        "final_train_loss": train_losses[-1],
        "final_val_loss": val_losses[-1],
        "runtime_min": parse_runtime_minutes(log_path),
    }


def collect_run_result(spec):
    log_path = Path(spec["log"])
    if not log_path.exists():
        return {"status": "missing log"}
    try:
        return {"status": "ready", **final_metrics(log_path)}
    except ValueError as exc:
        return {"status": str(exc)}


shakespeare_256_width_results = final_metrics("../logs/256_width.log")

pride_prejudice_run_results = {}
for run_name, spec in pride_prejudice_run_specs.items():
    pride_prejudice_run_results[run_name] = {
        "dataset": "Pride and Prejudice",
        "dataset_size_chars": 726900,
        "vocab_size": 91,
        "preprocessing": "Removed Project Gutenberg header, license text, and footer.",
        **spec,
        **collect_run_result(spec),
    }

available_pride_prejudice_results = {
    run_name: result
    for run_name, result in pride_prejudice_run_results.items()
    if result.get("status") == "ready"
}

pride_prejudice_results = pride_prejudice_run_results["Pride_and_Prejudice_256_Width"]

run_comparison = {
    "Shakespeare_256_width": {
        "dataset": "Tiny Shakespeare",
        **shakespeare_256_width_results,
    },
    **pride_prejudice_run_results,
}

run_comparison
```
