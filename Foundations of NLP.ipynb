### 4.2 Loss curves

Create a **Python script** that:

• Parses the training logs for each run (nanoGPT prints step, train loss, val loss to stdout).

• Plots **training loss vs. step** for all runs in a single figure (one line per run, with a legend).

• Plots **validation loss vs. step** in a second figure.

Save these as .png or .pdf files. **Tip:** Redirect output to a log file during training:

Because the notebook runs inside `nanoGPT/`, the plotting code below reads logs from the project-root `../logs/` directory and saves figures back to the assignment root.

```python
!python train.py ../config/train_shakespeare_char_baseline.py | tee ../logs/baseline.log
```

```python
# OUR PYTHON CODE
import sys
from pathlib import Path

sys.path.append("..")
from loss_curves import parse_log_file, plot_training_loss, plot_validation_loss

LOG_FILES = [
    "../logs/baseline.log",
    "../logs/low_lr.log",
    "../logs/2_layers.log",
    "../logs/4_layers.log",
    "../logs/128_width.log",
    "../logs/256_width.log",
    "../logs/64_context.log",
    "../logs/128_context.log",
    "../logs/1000_iters.log",
    "../logs/2500_iters.log",
]

all_runs = {}

for log_name in LOG_FILES:
    log_path = Path(log_name)

    if not log_path.exists():
        print(f"Skipping missing file: {log_path}")
        continue

    steps, train_losses, val_losses = parse_log_file(log_path)

    run_name = log_path.stem
    all_runs[run_name] = {
        "steps": steps,
        "train": train_losses,
        "val": val_losses,
    }
    
if not all_runs:
    raise FileNotFoundError("No valid log files were found.")

plot_training_loss(all_runs, output_path="../training_loss.png")
plot_validation_loss(all_runs, output_path="../validation_loss.png")
```

```python
# Parse final metrics from the Pride and Prejudice log and compare them with the Shakespeare 256-width run
import sys
import importlib

sys.path.append("..")
import loss_curves
importlib.reload(loss_curves)

from loss_curves import parse_log_file, parse_runtime_minutes


def final_metrics(log_name):
    log_path = Path(log_name)
    steps, train_losses, val_losses = parse_log_file(log_path)
    if not steps:
        raise ValueError(f"No step metrics found in {log_path}")
    return {
        "final_step": steps[-1],
        "final_train_loss": train_losses[-1],
        "final_val_loss": val_losses[-1],
        "runtime_min": parse_runtime_minutes(log_path),
    }

shakespeare_256_width_results = final_metrics("../logs/256_width.log")
pride_prejudice_metrics = final_metrics("../logs/pride_prejudice.log")

pride_prejudice_results = {
    "dataset": "Pride and Prejudice",
    "dataset_size_chars": 726900,
    "vocab_size": 91,
    "preprocessing": "Removed Project Gutenberg header, license text, and footer.",
    **pride_prejudice_metrics,
}

run_comparison = {
    "Shakespeare_256_width": {
        "dataset": "Tiny Shakespeare",
        **shakespeare_256_width_results,
    },
    "Pride_and_Prejudice": pride_prejudice_results,
}

run_comparison
```
