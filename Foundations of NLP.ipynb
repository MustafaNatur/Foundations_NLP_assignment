```python
# OUR PYTHON CODE
import sys
from pathlib import Path

sys.path.append("..")
from loss_curves import parse_log_file, plot_training_loss, plot_validation_loss

LOG_FILES = [
    "../logs/baseline.log",
    "../logs/low_lr.log",
    "../logs/2_layers.log",
    "../logs/4_layers.log",
    "../logs/128_width.log",
    "../logs/256_width.log",
    "../logs/64_context.log",
    "../logs/128_context.log",
    "../logs/1000_iters.log",
    "../logs/2500_iters.log",
]

all_runs = {}

for log_name in LOG_FILES:
    log_path = Path(log_name)

    if not log_path.exists():
        print(f"Skipping missing file: {log_path}")
        continue

    steps, train_losses, val_losses = parse_log_file(log_path)

    run_name = log_path.stem
    all_runs[run_name] = {
        "steps": steps,
        "train": train_losses,
        "val": val_losses,
    }
    
if not all_runs:
    raise FileNotFoundError("No valid log files were found.")

plot_training_loss(all_runs, output_path="../training_loss.png")
plot_validation_loss(all_runs, output_path="../validation_loss.png")
```

```python
# Parse final metrics from the retained Pride and Prejudice runs and compare them with the Shakespeare 256-width run
import sys
import importlib
from pathlib import Path

sys.path.append("..")
import loss_curves
importlib.reload(loss_curves)

from loss_curves import parse_log_file, parse_runtime_minutes

pride_prejudice_run_specs = {
    "Pride_and_Prejudice_Baseline": {
        "log": "../logs/pride_prejudice_baseline.log",
        "out_dir": "../Weights/out-pride-prejudice-baseline",
    },
    "Pride_and_Prejudice_64_Context": {
        "log": "../logs/pride_prejudice_64_context.log",
        "out_dir": "../Weights/out-pride-prejudice-64_context",
    },
    "Pride_and_Prejudice_256_Width": {
        "log": "../logs/pride_prejudice.log",
        "out_dir": "../Weights/out-pride-prejudice",
    },
}


def final_metrics(log_name):
    log_path = Path(log_name)
    steps, train_losses, val_losses = parse_log_file(log_path)
    if not steps:
        raise ValueError(f"No step metrics found in {log_path}")
    return {
        "final_step": steps[-1],
        "final_train_loss": train_losses[-1],
        "final_val_loss": val_losses[-1],
        "runtime_min": parse_runtime_minutes(log_path),
    }


def collect_run_result(spec):
    log_path = Path(spec["log"])
    if not log_path.exists():
        return {"status": "missing log"}
    try:
        return {"status": "ready", **final_metrics(log_path)}
    except ValueError as exc:
        return {"status": str(exc)}


shakespeare_256_width_results = final_metrics("../logs/256_width.log")

pride_prejudice_run_results = {}
for run_name, spec in pride_prejudice_run_specs.items():
    pride_prejudice_run_results[run_name] = {
        "dataset": "Pride and Prejudice",
        "dataset_size_chars": 726900,
        "vocab_size": 91,
        "preprocessing": "Removed Project Gutenberg header, license text, and footer.",
        "hardware": "Apple M4 via PyTorch MPS (reduced three-run, 1500-step bonus setup)",
        **spec,
        **collect_run_result(spec),
    }

available_pride_prejudice_results = {
    run_name: result
    for run_name, result in pride_prejudice_run_results.items()
    if result.get("status") == "ready"
}

pride_prejudice_results = pride_prejudice_run_results["Pride_and_Prejudice_256_Width"]

run_comparison = {
    "Shakespeare_256_width": {
        "dataset": "Tiny Shakespeare",
        **shakespeare_256_width_results,
    },
    **pride_prejudice_run_results,
}

run_comparison
```

#### Bonus A Loss Curves

Create the same two loss-curve plots as in Section 4.2, but only for the retained Pride and Prejudice bonus runs.

Because the notebook runs inside `nanoGPT/`, the plotting code below reads the bonus logs from the project-root `../logs/` directory and saves the figures back to the assignment root as `pride_prejudice_training_loss.png` and `pride_prejudice_validation_loss.png`.


```python
# Bonus A loss curves for the retained Pride and Prejudice runs
import sys
from pathlib import Path

sys.path.append("..")
from loss_curves import parse_log_file, plot_training_loss, plot_validation_loss

BONUS_LOG_FILES = {
    "baseline": "../logs/pride_prejudice_baseline.log",
    "64_context": "../logs/pride_prejudice_64_context.log",
    "256_width": "../logs/pride_prejudice.log",
}

bonus_runs = {}

for run_name, log_name in BONUS_LOG_FILES.items():
    log_path = Path(log_name)

    if not log_path.exists():
        print(f"Skipping missing file: {log_path}")
        continue

    steps, train_losses, val_losses = parse_log_file(log_path)
    if not steps:
        print(f"Skipping empty metrics file: {log_path}")
        continue

    bonus_runs[run_name] = {
        "steps": steps,
        "train": train_losses,
        "val": val_losses,
    }

if not bonus_runs:
    raise FileNotFoundError("No valid Pride and Prejudice bonus logs were found.")

plot_training_loss(bonus_runs, output_path="../pride_prejudice_training_loss.png")
plot_validation_loss(bonus_runs, output_path="../pride_prejudice_validation_loss.png")
print("Saved ../pride_prejudice_training_loss.png and ../pride_prejudice_validation_loss.png")
```
